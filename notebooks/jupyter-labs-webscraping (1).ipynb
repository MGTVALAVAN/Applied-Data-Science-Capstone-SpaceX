````python
import sys

import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
````

and we will provide some helper functions for you to process web scraped HTML table


````python
def date_time(table_cells):
    """
    This function returns the data and time from the HTML  table cell
    Input: the  element of a table data cell extracts extra row
    """
    return [data_time.strip() for data_time in list(table_cells.strings)][0:2]

def booster_version(table_cells):
    """
    This function returns the booster version from the HTML  table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=''.join([booster_version for i,booster_version in enumerate( table_cells.strings) if i%2==0][0:-1])
    return out

def landing_status(table_cells):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=[i for i in table_cells.strings][0]
    return out


def get_mass(table_cells):
    mass=unicodedata.normalize("NFKD", table_cells.text).strip()
    if mass:
        mass.find("kg")
        new_mass=mass[0:mass.find("kg")+2]
    else:
        new_mass=0
    return new_mass


def extract_column_from_header(row):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    if (row.br):
        row.br.extract()
    if row.a:
        row.a.extract()
    if row.sup:
        row.sup.extract()
        
    colunm_name = ' '.join(row.contents)
    
    # Filter the digit and empty names
    if not(colunm_name.strip().isdigit()):
        colunm_name = colunm_name.strip()
        return colunm_name    
````

### TASK 2: Extract all column/variable names from the HTML table header


Next, we want to collect all relevant column names from the HTML table header


Let's try to find all tables on the wiki page first. If you need to refresh your memory about `BeautifulSoup`, please check the external reference link towards the end of this lab


````python
import requests
from bs4 import BeautifulSoup

# Define URL
url = "https://en.wikipedia.org/wiki/List_of_Falcon_9_and_Falcon_Heavy_launches"

# Add User-Agent header (CRITICAL)
headers = {
    "User-Agent": "Mozilla/5.0"
}

# HTTP GET request
response = requests.get(url, headers=headers)

# Verify request
print("HTTP status code:", response.status_code)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Verify page title
print("Page title:", soup.title)

# Find tables
tables = soup.find_all("table")
print("Total tables found:", len(tables))

# Extract column names from the first wikitable
for table in tables:
    if table.get("class") and "wikitable" in table.get("class"):
        column_names = [th.text.strip() for th in table.find_all("th")]
        break

column_names
````

Next, we just need to iterate through the `<th>` elements and apply the provided `extract_column_from_header()` to extract column name one by one


````python
column_names = []

# Apply find_all() function with `th` element on first_launch_table
# Iterate each th element and apply the provided extract_column_from_header() to get a column name
# Append the Non-empty column name (`if name is not None and len(name) > 0`) into a list called column_names
# Select the target table (3rd table)
launch_table = tables[2]

# Extract all rows
rows = launch_table.find_all("tr")

# Initialize launch dictionary
launch_dict = {
    "Flight No.": [],
    "Date and time (UTC)": [],
    "Version, Booster": [],
    "Launch site": [],
    "Payload": [],
    "Payload mass": [],
    "Orbit": [],
    "Customer": [],
    "Launch outcome": []
}

# Loop through rows (skip header row)
for row in rows[1:]:
    cols = row.find_all("td")
    
    # Skip rows that don't have full data
    if len(cols) < 9:
        continue
    
    launch_dict["Flight No."].append(cols[0].text.strip())
    launch_dict["Date and time (UTC)"].append(cols[1].text.strip())
    launch_dict["Version, Booster"].append(cols[2].text.strip())
    launch_dict["Launch site"].append(cols[3].text.strip())
    launch_dict["Payload"].append(cols[4].text.strip())
    launch_dict["Payload mass"].append(cols[5].text.strip())
    launch_dict["Orbit"].append(cols[6].text.strip())
    launch_dict["Customer"].append(cols[7].text.strip())
    launch_dict["Launch outcome"].append(cols[8].text.strip())

# Sanity check: number of launches extracted
len(launch_dict["Flight No."])
````
